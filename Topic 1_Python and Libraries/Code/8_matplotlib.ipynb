```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
housing = pd.read_csv('housing.csv')
```

## 1. Histogram

```python
price = housing.median_house_value
price
```

```python
price.hist() # seems we didn't use matplotlib. Actually it is used. 
plt.show()
```

```python
plt.hist(price)
plt.grid()
plt.show()
```

```python
plt.hist(np.log(price))
plt.grid()
plt.show()
```

## 2. Scatter plot

```python
housing.head()
```

```python
plt.scatter(housing['longitude'], housing['latitude'])
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.show()
```

```python
plt.scatter(housing['longitude'], housing['latitude'], alpha=0.2)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.show()
```

```python
plt.scatter(housing['median_income'], housing['median_house_value'])
plt.show()
```

```python
plt.plot(housing['median_income'], housing['median_house_value'], 'o')
plt.show()
```

## 3. Line plot

```python
x = np.arange(10)
y = 2 * x + np.random.rand(10)*2

y2 = 2 * x + np.random.rand(10)*2
```

```python
plt.plot(x, y)
plt.show()
```

- line style
- marker
- color

```python
plt.plot(x, y, linestyle = 'dashed', marker = 'o', color='red')
plt.plot(x, y2, linestyle = 'dotted', marker = '^', color = 'blue')
plt.grid()
plt.show()
```
